# Image Compression with Singular Value Decomposition#

## Introduction

The Singular Value Decomposition (SVD) is a powerful linear algebra technique that can be used for image compression. By reducing the number of singular values used to reconstruct an image, we can significantly compress the image while retaining most of its visual quality.

## Key Concepts
1. **SVD in Linear Algebra**:
   - Any matrix \( A \) can be decomposed as:
     \[
     A = U \Sigma V^T
     \]
     - \( U \): Orthogonal matrix of left singular vectors.
     - \( \Sigma \): Diagonal matrix of singular values.
     - \( V^T \): Orthogonal matrix of right singular vectors.
   - The singular values in \( \Sigma \) represent the importance of each corresponding dimension in the data.

2. **Image Compression**:
   - Images are represented as matrices where each entry corresponds to pixel intensity.
   - Using fewer singular values (truncating \( \Sigma \)), we can approximate the original image with reduced storage requirements.

3. **Trade-off**:
   - More singular values -> Higher quality but larger size.
   - Fewer singular values -> Smaller size but lower quality.



## 1. Data

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

image = Image.open('/Users/stitchbleu/Desktop/inde-577-project/Data/leana.jpeg')
gray_image = image.convert('L')

image_array = np.array(gray_image)

plt.imshow(image_array, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

## 2. Application

```python
# Perform Singular Value Decomposition
U, S, VT = np.linalg.svd(image_array, full_matrices=False)

# Display singular values
plt.plot(S)
plt.title("Singular Values of the Image")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
```

## 3. Image Reconstruction with Truncated SVD

```python

def reconstruct_image(U, S, VT, k):
    S_k = np.diag(S[:k])
    U_k = U[:, :k]
    VT_k = VT[:k, :]
    return np.dot(U_k, np.dot(S_k, VT_k))

ks = [10, 50, 100, 200]  
for k in ks:
    reconstructed = reconstruct_image(U, S, VT, k)
    plt.imshow(reconstructed, cmap='gray')
    plt.title(f"Reconstructed Image with k={k}")
    plt.axis('off')
    plt.show()
```

## 4. Conclusion


The Singular Value Decomposition (SVD) technique demonstrates its power in image compression by effectively reducing the size of an image while maintaining visual quality. By truncating the number of singular values used in the reconstruction, we observe a trade-off between compression and image fidelity. With fewer singular values (e.g., \(k = 10\)), the image becomes blurry and loses details but achieves significant compression. As the number of singular values increases (e.g., \(k = 100\) or \(k = 200\)), the reconstructed image closely approximates the original, preserving essential features while still achieving a reduction in data storage requirements. This method highlights SVD's practical applications in storage optimization, transmission efficiency, and preprocessing for machine learning tasks, making it a fundamental tool in image processing and data science.
