# DBSCAN

## Introduction

 DBSCAN is a popular clustering algorithm used in unsupervised machine learning. 
 It identifies clusters of points in a dataset based on their density, making it particularly 
 useful for datasets with noise and varying cluster shapes.

 Key Concepts:
 1. **Core Points**: A point is a core point if it has at least `min_samples` points 
    (including itself) within a distance `eps`.
 2. **Border Points**: A point is a border point if it is within the `eps` neighborhood of a core point 
    but does not itself meet the criteria to be a core point.
 3. **Noise Points**: Points that are neither core nor border points are labeled as noise (outliers).

 DBSCAN Advantages:
 - Does not require the number of clusters to be predefined (unlike k-means).
 - Can discover arbitrarily shaped clusters.
 - Handles noise effectively.

DBSCAN Algorithm Steps
1. **Select a random unvisited point**:
    - If itâ€™s a core point, form a new cluster.
    - Otherwise, label it as noise.
2. **Expand the cluster**:
    - For each core point, recursively add its density-connected neighbors to the cluster.
3. **Repeat until all points are visited**.

Parameters:
- `eps`: The maximum distance between two points for one to be considered as part of the neighborhood.
- `min_samples`: The minimum number of points required to form a dense region (including the core point).

Applications of DBSCAN:
- Anomaly detection (e.g., fraud detection, sensor data anomalies).
- Geographic data analysis (e.g., identifying regions of high population density).
- Image segmentation and computer vision tasks.


## 1. Data

```python
import pandas as pd

data = pd.read_csv('/Users/stitchbleu/Desktop/inde-577-project/Data/day.csv')
data.head()
```

## 2. Application

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np

features = data[['temp', 'hum', 'windspeed']]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(scaled_features)

data['DBSCAN_Cluster'] = clusters

# temp and hum
plt.scatter(data['temp'], data['hum'], c=data['DBSCAN_Cluster'], cmap='viridis', s=10)
plt.title("DBSCAN Clustering: Temperature vs Humidity")
plt.xlabel("Temperature (normalized)")
plt.ylabel("Humidity (normalized)")
plt.colorbar(label="Cluster Label")
plt.show()

# temp and windspeed
plt.scatter(data['temp'], data['windspeed'], c=data['DBSCAN_Cluster'], cmap='viridis', s=10)
plt.title("DBSCAN Clustering: Temperature vs Windspeed")
plt.xlabel("Temperature (normalized)")
plt.ylabel("Windspeed (normalized)")
plt.colorbar(label="Cluster Label")
plt.show()
```

## 3.  DBSCAN with different parameters

```python
parameter_sets = [
    {"eps": 0.3, "min_samples": 5},
    {"eps": 0.5, "min_samples": 10},
    {"eps": 0.7, "min_samples": 5},
]

for params in parameter_sets:
    dbscan = DBSCAN(eps=params["eps"], min_samples=params["min_samples"])
    clusters = dbscan.fit_predict(scaled_features)
    data['DBSCAN_Cluster'] = clusters

    plt.figure(figsize=(8, 5))
    plt.scatter(data['temp'], data['hum'], c=data['DBSCAN_Cluster'], cmap='viridis', s=10)
    plt.title(f"DBSCAN Clustering (eps={params['eps']}, min_samples={params['min_samples']})")
    plt.xlabel("Temperature (normalized)")
    plt.ylabel("Humidity (normalized)")
    plt.colorbar(label="Cluster Label")
    plt.show()
```

## 4. conclusion

**DBSCAN** is a versatile clustering algorithm capable of identifying clusters based on density while effectively handling noise points. From the analysis with the bike-sharing dataset, the following insights were observed:

## 1. Parameter Sensitivity
- **`eps` (Neighborhood Radius)**: 
  - A smaller `eps` creates tighter, more localized clusters but increases the number of noise points.
  - A larger `eps` merges nearby clusters, reducing the number of distinct clusters but potentially overgeneralizing the data structure.
- **`min_samples` (Minimum Points in a Cluster)**:
  - Increasing `min_samples` makes DBSCAN more selective in forming clusters, often requiring denser regions.
  - This can help filter out small, insignificant clusters but may misclassify borderline points as noise.

## 2. Noise Detection
- DBSCAN effectively identifies noise points, which are labeled as `-1`. 
- The amount of noise detected depends on the chosen parameters and the dataset's density distribution.

## 3. Application Insights
- For the bike-sharing dataset, DBSCAN successfully grouped days with similar temperature, humidity, and windspeed characteristics.
- These clusters could help identify weather patterns influencing bike rentals or segment days for targeted analysis.

## 4. Trade-offs
- Choosing appropriate parameters (`eps` and `min_samples`) is critical and often requires domain knowledge or experimentation.
- Visualization of results is essential to understand the clustering behavior and evaluate noise classification.

---

## Recommendations
1. Experiment with a range of `eps` and `min_samples` values to find the optimal balance for your dataset.
2. Combine DBSCAN with domain expertise or additional preprocessing to ensure meaningful cluster formation.
3. Use visualization techniques to assess cluster quality and noise classification.

---

**Conclusion**: DBSCAN's flexibility and robustness make it a valuable tool for exploring datasets with varying cluster shapes and densities. It is particularly effective in handling noise and discovering patterns without requiring the number of clusters to be predefined.
