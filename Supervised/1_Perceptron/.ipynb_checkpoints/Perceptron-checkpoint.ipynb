# Perceptron

## Introduction
The perceptron is one of the simplest types of artificial neural networks, often used for binary classification tasks. It consists of a single linear decision boundary that separates two classes. The perceptron algorithm iteratively adjusts weights based on classification errors to find the optimal decision boundary. This notebook explores the perceptronâ€™s use in binary classification and evaluates its effectiveness on a sample dataset.

## 1. Data

```python

from ucimlrepo import fetch_ucirepo
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

heart_disease = fetch_ucirepo(id=45)  

data = heart_disease.data.features
data['target'] = heart_disease.data.targets

print("First few rows of the data:")
print(data.head())
print("\nData Information:")
data.info()

X = data.drop('target', axis=1)
y = data['target']

categorical_features = ['cp', 'thal', 'slope'] 
X = pd.get_dummies(X, columns=categorical_features, drop_first=True)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print("\nData is ready and has been split into training and testing sets.")
```

## 2. Define and train the model

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

perceptron_model = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
perceptron_model.fit(X_train, y_train)
y_pred = perceptron_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Perceptron Model Accuracy on Test Set: {accuracy:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

conf_matrix = confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=["0", "1", "2", "3", "4"])
fig, ax = plt.subplots(figsize=(8, 6))
disp.plot(cmap="Blues", ax=ax)
plt.title("Confusion Matrix for Perceptron Model")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

## 3.Train the Perceptron Model with Class Weights

```python
perceptron_model_balanced = Perceptron(max_iter=1000, tol=1e-3, random_state=42, class_weight='balanced')
perceptron_model_balanced.fit(X_train, y_train)

y_pred_balanced = perceptron_model_balanced.predict(X_test)

from sklearn.metrics import classification_report
print("Classification Report with Balanced Class Weights:")
print(classification_report(y_test, y_pred_balanced, zero_division=1))
```

## 4. SMOTE

```python
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import Perceptron
from sklearn.metrics import classification_report

smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

from collections import Counter
print("Class distribution after SMOTE:", Counter(y_train_smote))
perceptron_model_smote = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
perceptron_model_smote.fit(X_train_smote, y_train_smote)

y_pred_smote = perceptron_model_smote.predict(X_test)

print("\nClassification Report with SMOTE:")
print(classification_report(y_test, y_pred_smote, zero_division=1))
```

## 5. Visualization

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

conf_matrix = confusion_matrix(y_test, y_pred_smote)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=["0", "1", "2", "3", "4"])
fig, ax = plt.subplots(figsize=(8, 6))
disp.plot(cmap="Blues", ax=ax)
plt.title("Confusion Matrix for Perceptron Model with SMOTE")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

## PCA

```python
from sklearn.decomposition import PCA
import numpy as np

pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_smote)
X_test_pca = pca.transform(X_test)

perceptron_model_pca = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
perceptron_model_pca.fit(X_train_pca, y_train_smote)

x_min, x_max = X_train_pca[:, 0].min() - 1, X_train_pca[:, 0].max() + 1
y_min, y_max = X_train_pca[:, 1].min() - 1, X_train_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

Z = perceptron_model_pca.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 8))
plt.contourf(xx, yy, Z, alpha=0.3, cmap="coolwarm")
scatter = plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train_smote, cmap="coolwarm", edgecolor="k", s=20)
plt.legend(*scatter.legend_elements(), title="Classes")
plt.title("Decision Boundary with SMOTE on PCA-Reduced Data")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
